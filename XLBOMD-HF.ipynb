## XL-BOMD for simple HF
### From A. Niklasson's ML code

A simple conversion of the Matlab (ML) code of XL-BOMD for Hartree-Fock of H2. More to come. See:


A.M.N. Niklasson & M. Cawkwell J. Chem. Phys. 141, 164123 (2014)      
P. Souvatzis & A.M.N. Niklasson J. Chem. Phys. 140, 044117 (2014)   
P. Steneteg et al. Phys. Rev. B 82, 075110 (2010)            
A.M.N Niklasson Phys. Rev. Lett. 100, 123001 (2008)           

In XL-BOMD, the regular BO Lagrangian $\mathcal{L}^\mathrm{BO} $ is extended with a set of auxilliary (density or) wave functions $ \Phi = \{\phi_{nk} \}$ evolving in harmonic oscillations around the self-consistent (sc)
ground state wave functions $\Psi^{sc}(t)$:
\begin{equation}
\mathcal{L}^\mathrm{XBO}(\mathbf{R}, \mathbf{\dot{R}}, \mathbf{\Phi}, \mathbf{\dot{\Phi}}) = \mathcal{L}^\mathrm{BO} + \frac{\mu}{2} \sum_{nk} \int \left \vert \dot{\Phi}_{nk} \right \vert^2 d \mathbf{r} - \frac{\mu \omega^2}{2} \sum_{nk} \int \left \vert \Psi^{nc}_{nk}- \Phi_{nk} \right \vert^2 d \mathbf{r} 
\end{equation}
Where $\mu$ is a ficticious electron mass parameter, and $\omega$ is a frequency / curvature parameter for the harmonic potentials. 

```python
r = 1.40; v = 0.2;            # Initial position and velocity
c = 0.6; dr = 0.001;          # Mixing and finite difference step
Nocc = 1;
```

Define functions:

```python
from math import erf
from numpy import pi, sqrt, exp

''' basis '''
alpha1 = np.array([13.00773, 1.962079, 0.444529, 0.1219492]);
alpha = np.append(alpha1, alpha1)
asize = np.shape(alpha)

def overlap_kinetic(r):
    S = np.zeros((8, 8))
    T = np.zeros((8, 8))
    ra = np.array([0., 0., 0., 0., r, r, r, r]) # the location of the atom with the ith basisfunction
    for i in range(8): # run over all basis functions
        for j in range(0, i + 1): # do not run over all basis functions to exploit symmetry
            # intermediate variables
            a_s = alpha[i] + alpha[j]
            ap = alpha[i] * alpha[j]
            rat = ap/a_s
            # location of the gaussian resulting from product of
            # the 2 gaussian basis functions
            rp = (alpha[i] * ra[i] + alpha[j] * ra[j]) / a_s
            S[i, j] = (np.pi / a_s)**1.5 * np.exp(-rat * (ra[i] - ra[j])**2)
            S[j, i] = S[i, j] # using symmetry
            T[i, j] = 0.5 * rat * (6 - 4 * rat *(ra[i] - ra[j])**2) * S[i, j]
            T[j, i] = T[i, j] # using symmetry
            
    return np.matrix(S), np.matrix(T)

def external(r):
    ra = np.array([0., 0., 0., 0., r, r, r, r]) # the location of the atom with the ith basisfunction
    Z = 1.;
    A = np.zeros((8,8))
    for i in range(8): # run over all basis functions
        for j in range(0, i + 1): # do not run over all basis functions to exploit symmetry
            # intermediate variables
            a_s = alpha[i] + alpha[j]
            a_p = alpha[i] * alpha[j]
            rat = a_p / a_s
            # location of the gaussian resulting from product of
            # the 2 gaussian basis functions
            rp = (alpha[i] * ra[i] + alpha[j] * ra[j]) / a_s
            if rp == 0:
                F0 = 1 + np.sqrt(np.pi) / 2 * erf(np.sqrt(a_s * (rp - r)**2)) / np.sqrt(a_s * (rp - r)**2)
            elif rp == r:
                F0 = np.sqrt(np.pi) / 2 * erf(np.sqrt(a_s * (rp - 0)**2)) / np.sqrt(a_s * (rp - 0)**2) + 1
            else:
                F0 = np.sqrt(np.pi)/2 * (erf(np.sqrt(a_s * (rp - 0)**2)) / 
                                         np.sqrt(a_s * (rp - 0)**2) + erf(np.sqrt(a_s * (rp - r)**2)) / 
                                         np.sqrt(a_s * (rp - r)**2))

            A[i, j] = -2 *np.pi * Z / a_s * np.exp(-rat * (ra[i] - ra[j])**2) * F0
            A[j, i] = A[i, j] # using symmetry
            
    return A

def twoelint(r):
    ra = np.array([0., 0., 0., 0., r, r, r, r]) # the location of the atom with the ith basisfunction
    g = np.zeros((8, 8, 8, 8))
    for i in range(8): # run over all basis functions
        for j in range(0, i + 1): # do not run over all basis functions to exploit symmetry
            as1=alpha[i]+alpha[j];
            ap1=alpha[i]*alpha[j];
            rp=(alpha[i]*ra[i]+alpha[j]*ra[j])/as1;
            for k in range(i): #k=1:i-1 #do not run over all basis functions to exploit symmetry
                for l in range(k + 1):  #l=1:k # do not run over all basis functions to exploit symmetry
                    as2=alpha[k]+alpha[l];
                    ap2=alpha[k]*alpha[l];
                    rq=(alpha[k]*ra[k]+alpha[l]*ra[l])/as2;
                    if ((rp-rq)==0):
                        F0=1;
                    else:
                        F0=sqrt(pi)/2*erf(sqrt(as1*as2*(rp-rq)**2/(as1+as2)))/sqrt(as1*as2*(rp-rq)**2/(as1+as2));
                        
                    g[i,j,k,l]=2*pi**2.5/as1/as2/(as1+as2)**0.5*exp(-ap1*(ra[i]-ra[j])**2/as1-ap2*(ra[k]-ra[l])**2/as2)*F0;
                    g[k,l,i,j]=g[i,j,k,l] # using symmetry
                    g[j,i,k,l]=g[i,j,k,l]
                    g[i,j,l,k]=g[i,j,k,l]
                    g[j,i,l,k]=g[i,j,k,l]
                    g[k,l,j,i]=g[i,j,k,l]
                    g[l,k,i,j]=g[i,j,k,l]
                    g[l,k,j,i]=g[i,j,k,l]

                k=i;
                for l in range(j + 1): #l=1:j
                    as2=alpha[k]+alpha[l];
                    ap2=alpha[k]*alpha[l];
                    rq=(alpha[k]*ra[k]+alpha[l]*ra[l])/as2;
                    if ((rp-rq)==0):
                        F0=1;
                    else:
                        F0=sqrt(pi)/2*erf(sqrt(as1*as2*(rp-rq)**2/(as1+as2)))/sqrt(as1*as2*(rp-rq)**2/(as1+as2));

                    g[i,j,k,l]=2*pi**2.5/as1/as2/(as1+as2)**0.5*exp(-ap1*(ra[i]-ra[j])**2/as1-ap2*(ra[k]-ra[l])**2/as2)*F0;
                    g[k,l,i,j]=g[i,j,k,l] # using symmetry
                    g[j,i,k,l]=g[i,j,k,l] 
                    g[i,j,l,k]=g[i,j,k,l] 
                    g[j,i,l,k]=g[i,j,k,l] 
                    g[k,l,j,i]=g[i,j,k,l] 
                    g[l,k,i,j]=g[i,j,k,l] 
                    g[l,k,j,i]=g[i,j,k,l] 

    return g

def J_K(g, P):
    ''' Coloumb (J) and exchange (K)'''
    JK = np.zeros((8,8))
    for i in range(8):
        for j in range(8):
            J = 0
            for k in range(8):
                for l in range(8):
                    J = J + P[k, l] * g[i, j, k, l]
            JK[i, j] = J
    return np.matrix(JK)

def SP2(H0, Ne):
    ''' https://github.com/exmatex/CoSP2 and 
        https://arxiv.org/pdf/1302.7292.pdf
        
        Fancy way of getting density matrix ...'''
    N = np.max(np.shape(H0))
    I = np.eye(N);
    hN = np.max(np.linalg.eigvals(H0)) # Approximate upper bound, should be estimated by Gersgorin
    h1 = np.min(np.linalg.eigvals(H0)) # Approximate lower bound, should be estimated by Gersgorin
    X0 = (hN * I - H0) / (hN - h1); # X0 = F0(H), the first step in the recursion
    trX0 = np.trace(X0);
    IdmpErr_2 = 0; IdmpErr_1 = 0; IdmpErr_0 = 0; # For convergence error schecks
    Err = 1
    i = 0
    while Err > 0:
        i = i + 1
        X0_2 = X0 * X0
        trX0_2 = np.trace(X0_2)
        if abs(trX0_2 - Ne) < abs(2 * trX0 - trX0_2-Ne):
            sgm = -1
        else:
            sgm = 1

        trX0_old = trX0
        X0 = X0 + sgm * (X0 - X0_2)
        trX0 = np.trace(X0)
        IdmpErr_2 = IdmpErr_1
        IdmpErr_1 = IdmpErr_0
        IdmpErr_0 = abs(trX0-trX0_old)
        if i > 12:                        # Minimum number of steps
            if IdmpErr_0 < 1e-2:          # Minimum Idempotency Error
                if IdmpErr_0 > IdmpErr_2: # Worse after 2 iteration
                    Err = 0
        if i > 100:
            Err = 0
            PROBLEMS = 1   ### What ? 
            IdemptencyError = IdmpErr_0
    return X0


def dOverlap_dKinetic(r, dr):
    S, T = overlap_kinetic(r)
    Sp, Tp = overlap_kinetic(r + dr)
    Sp2, Tp2 = overlap_kinetic(r + 2 * dr)
    Sm, Tm = overlap_kinetic(r - dr)
    Sm2, Tm2 = overlap_kinetic(r - 2 * dr)
    dS = (-Sp2 + 8 * Sp - 8 * Sm + Sm2) / (12 * dr)
    dT = (-Tp2 + 8 * Tp - 8 * Tm + Tm2) / (12 * dr)
    
    return np.matrix(dS), np.matrix(dT)

def dExternal(r, dr):
    A = external(r)
    Ap = external(r + dr)
    Ap2 = external(r + 2 * dr)
    Am = external(r - dr)
    Am2 = external(r - 2 * dr)
    dA = (-Ap2 + 8 * Ap - 8 * Am + Am2) / (12 * dr)
    
    return np.matrix(dA)

def dTwoElInt(r, dr):
    g = twoelint(r)
    gp = twoelint(r + dr)
    gp2 = twoelint(r + 2 * dr)
    gm = twoelint(r - dr)
    gm2 = twoelint(r - 2 * dr)
    dg = (-gp2 + 8 * gp - 8 * gm + gm2) / (12 * dr)
    
    return dg
```

Charge independent integrals and potentials

```python
from scipy.linalg import fractional_matrix_power as fmp
NNr = 1/r;

S, T =  overlap_kinetic(r)  # same as ML

A = np.matrix(external(r))  # same as ML 
g = twoelint(r)             # same as ML within 1e-12 (probably from exporting to ML for comparison)
Z = np.matrix(fmp(S, -0.5)) # Inverse overlap factor
I = np.matrix(np.eye(8))
P = Z * I * Z.T / 8.        # Initial guess of density matrix  (same as ML)
```

SCF optimization P = SCF[P,r,n,c]

```python
Pold = P.copy()
change = np.zeros(100)
D_norm = np.zeros(100)

for i in range(100):  ######################################## IS THIS PHASE ALIGNED ? Does it need to be? Re: PRB82 paper
    F = T + A + J_K(g, P)                 # Fockian
    D = Z * SP2(Z.T * F * Z, Nocc) * Z.T  # Calculate Density Matrix 
    P = c * D + (1 - c) * P;              # Simple linear mixing - Final P is the same as in ML.

    # just me messing around and looking at stuff
    Pdiff = P - Pold
    change[i] = np.linalg.norm(Pdiff)
    Pold = P.copy()
    D_norm[i] = np.linalg.norm(D)
```

```python
%matplotlib inline
fig, ax = plt.subplots(2 ,1)
ax[0].plot(range(100), change)
ax[1].plot(range(100), D_norm);  # So things are converging. ... 
```

Calculate initial potential energy U[P,r]

```python
JK = J_K(g, P); F = T + A + JK          # Fockian F[P]
D = Z * SP2(Z.T * F * Z, Nocc)* Z.T     # Calculate Density Matrix from F[P]

# U = 2*trace((T+A)*D) + trace(J_K(g,D)*D) + NNr;                      # Regular potential not used
U = 2 * np.trace((T + A) * D) + np.trace(JK * (2 * D - P)) + NNr;      # Shadow potential
print U  # Same as ML
```

And forces: Force[P,r]

```python
dS, dT = dOverlap_dKinetic(r, dr) 
dA = dExternal(r, dr)
dg = dTwoElInt(r, dr) 
dJK = J_K(dg, P)
dNNr = -1/r**2
Pulay = -2 * np.trace(D * F * D * dS)

## dUdR = 2*trace((dT+dA)*D) + trace(J_K(dg,D)*D) + dNNr + Pulay; ## Regular force not used
dUdR = 2*np.trace((dT+dA)*D) + np.trace(dJK*(2*D-P)) + dNNr + Pulay
print dUdR  
```

## XL BOMD loop

```python
## Initial BC for P_n and coefficients for the modified Verlet scheme
# BC = boundary conditions? 

P = D*S  # Density matrix * overlap matrix. 
P_0 = D*S
P_1 = D*S 
P_2 = D*S 
P_3 = D*S 
P_4 = D*S 
P_5 = D*S

#################### Parameters for the weak external dissipative electronic force. Is this an "electronic thermostat"? 
C0 = -6   # Coefficients from table 1 in PRB 82, 2010
C1 = 14 
C2 = -8 
C3 = -3 
C4 = 4  
C5 = -1

kappa = 1.82 # this is a dt^2 * omega^2 that corresponds to the Cn parameters above. See PRB 82
alpha_md = 0.018 # Also from table one 
# all these values correspond to 'K = 5'. What does that mean?!?

######################


cc = 0.80  #% Scaled delta function approximation of Kernel coefficient cc = [0,1]
dt = 0.25  # Integration time step
dt = 1.0  # Integration time step
M_init = 6 # Extra SFC convergence the first ~10 steps for smooth initialization

tsteps = 500

Energy = np.zeros(tsteps + 1)
time = np.zeros(tsteps + 1)
RR = np.zeros(tsteps + 1)
vv = np.zeros(tsteps + 1)

for i in range(1, tsteps + 1):  ##### MAIN MD LOOP  ####
    ## Check total energy, i.e. shadow Hamiltonian constant of motion
    Energy[i] = 0.5*v**2 + U
    time[i] = (i-1)*dt 
    RR[i] = r
    vv[i] = v 
    
    if i % 10 == 0:
        print 'Step: {0:3d}, Total Energy: {1:7.5f}'.format(i, 0.5*v**2 + U)

    v = v - dt * dUdR/2.           # Update velocity
    r = r + dt * v                 # Update position

    ## Update charge-independent integrals and potentials for new r
    S_0 = S
    SI0 = S**(-1)  # this works because S is already np.matrix
    S, T = overlap_kinetic(r)
    A = np.matrix(external(r))
    g = twoelint(r)
    Z = np.matrix(fmp(S, -0.5)) 
    NNr = 1/r;

    ## Integrate the XL-BOMD equations of motion, P'' = w^2(D-P), with modified Verlet, i.e.  
    ## P(t+dt) = 2P(t)-P(t-dt)+dt^2*P''(t) + Weak Dissipation
    if i <= M_init:
        P_0 = D*S_0 # Use modified initial integration with high dissipation
        
    P = 2*P_0 - P_1 + cc*kappa*(D*S_0 - P_0) + alpha_md*(C0*P_0 + C1*P_1 + C2*P_2 + C3*P_3 + C4*P_4 + C5*P_5)

    P_5 = P_4; P_4 = P_3; P_3 = P_2; P_2 = P_1; P_1 = P_0; P_0 = P; # Shift values
    P = P*S**(-1);

    Nr_SCF_It = 0;                   # Set Nr_SCF_It > 0 for "exact" BOMD
    if i < M_init: 
        Nr_SCF_It = 4                # Use SCF initially 

    ## SCF optimization P = SCF[P,n,r,c]
    P0 = P;
    for k in range(Nr_SCF_It): 
        F = T + A + J_K(g, P)               # Fockian
        D = Z * SP2(Z.T * F * Z, Nocc)*Z.T  # Calculate Density Matrix
        P = c * D + (1 - c) * P             # Simple linear mixing


    ## Shadow Hamiltonian U of XL-BOMD in 0-SCF limit {U[P,r]}
    JK = J_K(g, P)
    F = T + A + JK                              # Fockian F[P]
    D = Z * SP2(Z.T * F * Z, Nocc) * Z.T        # Calculate Density Matrix from F[P]
    U = 2 * np.trace((T + A) * D) + np.trace(JK * (2 * D - P)) + NNr # Shadow Hamiltonian 

    ## with exact force dU/dr {Force[P,r]}
    dS, dT = dOverlap_dKinetic(r,dr)
    dA = dExternal(r,dr)
    dg = dTwoElInt(r,dr)
    dJK = J_K(dg,P)
    dNNr = -1/r**2
    Pulay = -2 * np.trace(D * F * D * dS)
    dUdR_reg = 2*np.trace((dT + dA) * D) + np.trace(J_K(dg, D) * D) + dNNr + Pulay ## Regular force not used
    dUdR = 2*np.trace((dT + dA) * D) + np.trace(dJK * (2 * D - P)) + dNNr + Pulay

    v = v - dt * dUdR / 2.          # Update velocity
```
